# Election data EDA and cleaning

```python
import pandas as pd
```

```python
path = "../data/raw/countypres_2000-2024.csv"
elections = pd.read_csv(path, dtype={'county_fips': str})
```

```python
elections.head()
```

```python
elections = [~elections['state'].isin(['ALASKA', 'HAWAII'])]
```

```python
elections[elections['state'] == 'ALASKA']
```

```python
elections[(elections['year'] == 2024) & (elections['state_po'] == 'WI') & (elections['county_fips'] == '55001')]
```

```python
# pad FIPS to 5 digits (restore leading zeros)
elections['county_fips'] = elections['county_fips'].str.zfill(5)
```

```python
# filter data to just 2024 and the two major parties
e24 = elections[
    (elections['year'] == 2024) &
    (elections['party'].isin(['REPUBLICAN', 'DEMOCRAT']))
].copy()
```

```python
e24
```

```python
e24.info()
```

```python
e24.describe()
```

```python
# for each county: if TOTAL rows exist, keep only those rows
# otherwise, keep all rows
def resolve_county(group):
    if 'TOTAL' in group['mode'].values:
        return group[group['mode'] == 'TOTAL']
    return group

e24 = e24.groupby('county_fips', group_keys=False).apply(resolve_county)
```

```python
e24
```

Fix known issues with county FIPS codes

```python
# drop Kansas City, Missouri: reported separately from its 4 counties, can't cleanly join to utility/climate data
e24 = e24[e24['county_fips'] != '36000']

# Shannon County, SD was renamed to Oglala Lakota County in 2015
# its fips code was changed from  46113 to 46102
e24.loc[e24['county_fips'] == '46113', 'county_fips'] = '46102'

# drop Alaska: reports presidential returns by state house district, not by county, so won't align with other data
e24 = e24[e24['state'] != 'ALASKA']

# drop rows with missing FIPS
e24 = e24.dropna(subset=['county_fips'])
```

```python
# pivot to wide format - one row per county
pivoted = e24.pivot_table(
    index=['county_fips', 'state', 'county_name'],
    columns='party',
    values='candidatevotes',
    aggfunc='sum' # only multiple rows if there is no row with TOTAL as the mode
).reset_index()
```

```python
# attach total votes 
# max handles cases where totalvotes repeats across broken-out mode rows; sum would double-count)
totals = e24.groupby('county_fips')['totalvotes'].max().reset_index() # totalvotes is the same for all row for a county
pivoted = pivoted.merge(totals, on='county_fips')
```

```python
totals
```

```python
# now has a totalvotes columns
pivoted
```

```python
# calculate each party's share of the vote
pivoted['rep_share'] = pivoted['REPUBLICAN'] / pivoted['totalvotes']
pivoted['dem_share'] = pivoted['DEMOCRAT']   / pivoted['totalvotes']
pivoted['rep_margin'] = pivoted['rep_share'] - pivoted['dem_share']
```

```python
print(f"Final county count: {len(pivoted)}")
print(f"States covered: {pivoted['state'].nunique()}")
print("\nVote share summary:")
print(pivoted[['rep_share', 'dem_share', 'rep_margin']].describe())
```

```python
# flag any counties where major parties sum to >1 to check for potential double counting
suspect = pivoted[(pivoted['rep_share'] + pivoted['dem_share']) > 1.01]
if len(suspect) > 0:
    print(f"Warning: {len(suspect)} counties have rep+dem share > 1:")
    print(suspect[['state', 'county_name', 'rep_share', 'dem_share']])
else:
    print("\nAll counties have valid vote shares.")
```

```python
pivoted.to_csv('../data/processed/elections_2024_county.csv', index=False)
```
