# Consumption data cleaning

Dataset: Table 5A - https://www.eia.gov/electricity/sales_revenue_price/

This dataset provides average monthly consumption (kWh) of electricity for each state. We will use this to predict monthly utility bills.

```python
import pandas as pd
```

```python
df = pd.read_excel("../data/raw/table_5A.xlsx")
```

```python
df
```

```python
df.columns = ['state', 'num_customers', 'avg_consumption', 'avg_price', 'avg_bill']
```

```python
df
```

```python
df.drop(index=[0, 1], inplace=True)
```

```python
df
```

```python
df = df[['state', 'avg_consumption']]
```

```python
df
```

```python
df['state'].unique()
```

```python
# drop Hawaii, Alaska, and regions
df = df[~df['state'].isin(['Hawaii', 'Alaska', 'New England', 'Middle Atlantic', 'South Atlantic', 'East North Central', 'West North Central', 'District of Columbia', 'West South Central', 'East South Central', 'Mountain', 'Pacific Contiguous', 'Pacific Noncontiguous', 'U.S. Total'])].reset_index(drop=True)
```

```python
df['state'].unique()
```

Now have the contiguous United States

```python
df['state'].nunique()
```

```python
df
```

```python
df = df.sort_values('state')
```

```python
df
```

```python
df.to_csv("../data/processed/consumption.csv", index=False)
```
