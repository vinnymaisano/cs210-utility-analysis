In the dataset, the state code, which is the first two digits of climate_division_code, is determined by the alphabetical order of the two-letter state codes. However, actual FIPS codes use a different mapping of states to state codes. state_lookup will map the NOAA state code to the FIPS state code. This will allow us to merge this dataset with the election and utility datasets.

The mapping from the state codes used in the data to FIPS state codes can be viewed in the state code table in the documentation: https://www.ncei.noaa.gov/pub/data/cirs/climdiv/county-readme.txt

```python
state_lookup = {
    "01": "01",  # AL
    "02": "04",  # AZ
    "03": "05",  # AR
    "04": "06",  # CA
    "05": "08",  # CO
    "06": "09",  # CT
    "07": "10",  # DE
    "08": "12",  # FL
    "09": "13",  # GA
    "10": "16",  # ID
    "11": "17",  # IL
    "12": "18",  # IN
    "13": "19",  # IA
    "14": "20",  # KS
    "15": "21",  # KY
    "16": "22",  # LA
    "17": "23",  # ME
    "18": "24",  # MD
    "19": "25",  # MA
    "20": "26",  # MI
    "21": "27",  # MN
    "22": "28",  # MS
    "23": "29",  # MO
    "24": "30",  # MT
    "25": "31",  # NE
    "26": "32",  # NV
    "27": "33",  # NH
    "28": "34",  # NJ
    "29": "35",  # NM
    "30": "36",  # NY
    "31": "37",  # NC
    "32": "38",  # ND
    "33": "39",  # OH
    "34": "40",  # OK
    "35": "41",  # OR
    "36": "42",  # PA
    "37": "44",  # RI
    "38": "45",  # SC
    "39": "46",  # SD
    "40": "47",  # TN
    "41": "48",  # TX
    "42": "49",  # UT
    "43": "50",  # VT
    "44": "51",  # VA
    "45": "53",  # WA
    "46": "54",  # WV
    "47": "55",  # WI
    "48": "56",  # WY
}
```

Map the NOAA state codes to FIPS state codes, then reconstruct the county FIPs code by appending the 3-digit county code to the state code

```python
cdd['state_fips'] = cdd['noaa_state_code'].map(state_lookup)
cdd['county_fips'] = cdd['state_fips'] + cdd['county_3digit']
```

```python
cdd
```

Taking the sum of cooling degree days over the year will give an indicator of how much demand there is for heat in each county over the course of the year.

```python
month_cols = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
```

```python
cdd['cdd_annual'] = cdd[month_cols].sum(axis=1)
```

```python
cdd
```

In the end, just need the county code and the number of cooling degree days (CDDs)

```python
cdd = cdd[['county_fips', 'cdd_annual']]
```

We now have the total cooling degree days in 2024 for each county.

We will now use the function load_climdiv to perform the same operations to both the CDD and HDD files.

```python
# returns a dataframe with columns county_fips and cdd/hdd_annual
def load_climdiv(path, year):
    cols = ['climate_division_code', 'jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

    df = pd.read_csv(
        path,
        sep=r'\s+',
        header=None,
        names=cols,
        dtype={"climate_division_code": str}, 
    )

    # break up the division code into state, county, and year
    df['noaa_state_code'] = df['climate_division_code'].str[:2] # state code is first 2 digits
    df['county_3digit'] = df['climate_division_code'].str[2:5] # county code is next 3 digits
    df['year'] = df['climate_division_code'].str[-4:] # year is last 4 digits

    # re-order the columns, remove climate_division_code
    df = df[['noaa_state_code', 'county_3digit', 'year'] + list(df.columns)[1:-3]]

    # only want to observe year
    df = df[df['year'] == year].reset_index(drop=True)

    df['state_fips'] = df['noaa_state_code'].map(state_lookup)
    df['county_fips'] = df['state_fips'] + df['county_3digit']

    month_cols = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

    name = "cdd_annual" if "cdd" in path else "hdd_annual"
    df[name] = df[month_cols].sum(axis=1).astype(int)
    df = df[['county_fips', name]]
    return df
```

```python
cdd = load_climdiv("../data/raw/climdiv-cddccy-v1.0.0-20260406.txt", "2024")
hdd = load_climdiv("../data/raw/climdiv-hddccy-v1.0.0-20260406.txt", "2024")
```

```python
cdd
```

```python
hdd
```

The combined data will now hold the number of cooling and heating degree days for each county in the year 202.

```python
combined = pd.merge(cdd, hdd, on='county_fips', how='outer', validate='one_to_one')
```
