<h1>Analysis & models</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../data/processed/utility-climate-election.csv", 
                 dtype={"county_fips": str})

print(f"Shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
df.head()
```

```python
df.groupby("state")["avg_res_rate"].mean().sort_values(ascending=False)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram
axes[0].hist(df["avg_res_rate"], bins=60, edgecolor="white")
axes[0].axvline(df["avg_res_rate"].mean(), color="red", linestyle="--", 
                label=f"Mean: ${df['avg_res_rate'].mean():.3f}")
axes[0].axvline(df["avg_res_rate"].median(), color="green", linestyle="--",
                label=f"Median: ${df['avg_res_rate'].median():.3f}")
axes[0].set_xlabel("Residential rate ($/kWh)")
axes[0].set_ylabel("Number of counties")
axes[0].set_title("Distribution of county-level residential rates")
axes[0].legend()

# Boxplot to see outliers
axes[1].boxplot(df["avg_res_rate"], vert=False)
axes[1].set_xlabel("Residential rate ($/kWh)")
axes[1].set_title("Boxplot")
axes[1].set_yticks([])

plt.tight_layout()
plt.show()

print(df["avg_res_rate"].describe())
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

axes[0][0].hist(df["cdd_annual"], bins=50, edgecolor="white")
axes[0][0].set_title("Cooling Degree Days (annual)")
axes[0][0].set_xlabel("CDD")

axes[0][1].hist(df["hdd_annual"], bins=50, edgecolor="white")
axes[0][1].set_title("Heating Degree Days (annual)")
axes[0][1].set_xlabel("HDD")

axes[1][0].hist(df["rep_share"], bins=50, edgecolor="white")
axes[1][0].set_title("Republican vote share (2024)")
axes[1][0].set_xlabel("Rep share")

axes[1][1].hist(df["dem_share"], bins=50, edgecolor="white")
axes[1][1].set_title("Democrat vote share (2024)")
axes[1][1].set_xlabel("Dem share")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

axes[0].scatter(df["cdd_annual"], df["avg_res_rate"], alpha=0.3, s=10)
axes[0].set_xlabel("CDD (annual)")
axes[0].set_ylabel("Residential rate ($/kWh)")
axes[0].set_title("Rate vs Cooling Degree Days")

axes[1].scatter(df["hdd_annual"], df["avg_res_rate"], alpha=0.3, s=10)
axes[1].set_xlabel("HDD (annual)")
axes[1].set_ylabel("Residential rate ($/kWh)")
axes[1].set_title("Rate vs Heating Degree Days")

axes[2].scatter(df["rep_share"], df["avg_res_rate"], alpha=0.3, s=10)
axes[2].set_xlabel("Republican share")
axes[2].set_ylabel("Residential rate ($/kWh)")
axes[2].set_title("Rate vs Political lean")

plt.tight_layout()
plt.show()
```

```python
state_summary = (
    df.groupby("state")
    .agg(
        n_counties=("county_fips", "count"),
        mean_rate=("avg_res_rate", "mean"),
        std_rate=("avg_res_rate", "std"),
        mean_cdd=("cdd_annual", "mean"),
        mean_hdd=("hdd_annual", "mean"),
        mean_rep_share=("rep_share", "mean"),
    )
    .sort_values("mean_rate", ascending=False)
)
print(state_summary)
```

```python
# Calculate variance components
total_var = df["avg_res_rate"].var()
within_state_var = df.groupby("state")["avg_res_rate"].var().mean()
between_state_var = df.groupby("state")["avg_res_rate"].mean().var()

print(f"Total variance:           {total_var:.6f}")
print(f"Within-state (avg):       {within_state_var:.6f}")
print(f"Between-state:            {between_state_var:.6f}")
print(f"\nFraction between-state: {between_state_var / total_var:.1%}")
```
